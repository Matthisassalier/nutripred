# Question SQL

```python
from sqlalchemy import create_engine

# load dataset
df_sql = pd.read_csv('data.csv', delimiter="|")

# database connection
hostname="127.0.0.1"
dbname="openfoodfacts"
uname="root"
pwd="root"

# create SQLAlchemy engine to connect to MySQL Database
engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}".format(host=hostname, db=dbname, user=uname, pw=pwd))
connection = engine.connect() 

# # export dataframe to sql table                                   
# df_sql.to_sql('products', engine, index=False)
```

**Nombre de produits par nutriscore ?** 

```python
df_sql_nutriscore = pd.read_sql(
    "select nutriscore_grade, count(product_name) from products group by nutriscore_grade;", con=connection)
df_sql_nutriscore
```

**Quel est le pourcentage de représentation associé?**    

```python
df_sql_nutriscore_pour = pd.read_sql(
    "select nutriscore_grade, count(product_name) * 100.0 / sum(count(product_name)) over() from products group by nutriscore_grade;", con=connection)
df_sql_nutriscore_pour
```

**Les 4 valeurs possibles du Nova_Group.**

```python
df_sql_nutriscore_NG = pd.read_sql(
    "SELECT Nova_Group FROM products GROUP BY Nova_Group;", con=connection)
df_sql_nutriscore_NG
```

**Quel est la moyenne des fibres par nutriscore ?**

```python
df_sql_nutriscore_fiber = pd.read_sql(
    "SELECT nutriscore_grade ,AVG(fiber_100g) FROM products GROUP BY nutriscore_grade;", con=connection)
df_sql_nutriscore_fiber
```

**Quels sont le TOP10 des pnns_groups_2 qui comportent le plus d'additifs ?**

```python
df_sql_nutriscore_PNNS = pd.read_sql(
    "SELECT pnns_groups_2,additives_n FROM products GROUP BY additives_n ORDER BY max(additives_n) DESC LIMIT 10;", con=connection)
df_sql_nutriscore_PNNS
```

**Afficher le taux de sucre/fat moyen et médian des produits avec un nutriscore A ou B**

```python
df_sql_nutriscore_sugar_avg = pd.read_sql(
    "SELECT product_name, AVG(sugars_100g / fat_100g), nutriscore_grade FROM products WHERE nutriscore_grade = 'a' OR nutriscore_grade = 'b' GROUP BY product_name;", con=connection)
df_sql_nutriscore_sugar_avg
```

```python
df_sql_nutriscore_sugar_md = pd.read_sql(
    "SELECT product_name, nutriscore_grade, sugars_100g/fat_100g as t1 from products where nutriscore_grade = 'a' or nutriscore_grade ='b' GROUP BY product_name HAVING SUM(SIGN(1-SIGN(t1-t1))) = (COUNT(*)+1)/2;", con=connection)
df_sql_nutriscore_sugar_md
```

**Combien avons-nous de produits qui ont un taux de carbohydrates > aux sucres ?** 

```python
df_sql_nutriscore_carbo = pd.read_sql(
    "select count(product_name) FROM products where (carbohydrates_100g)>(sugars_100g);", con=connection)
df_sql_nutriscore_carbo
```
